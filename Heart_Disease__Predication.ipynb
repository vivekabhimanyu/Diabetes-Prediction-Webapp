Importing required modulus in the program

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Collecting data from Heart disease dataset 

```python
heart_disease_data = pd.read_csv("heart_disease_data.csv")
```

```python
# printing top 5 rows
heart_disease_data.head()
```

```python
heart_disease_data.groupby('age').max()
```

```python
# printing last five rows
heart_disease_data.tail()
```

```python
# total rows in dataset
heart_disease_data.shape
```

```python
# dataset information
heart_disease_data.info()
```

```python
heart_disease_data.describe() # it results total count ,its means,std , min value ,max value of each column
```

```python
# distribution of target value
heart_disease_data['target'].value_counts()
```

if target value is 1 then patient has heart disease 
if target value is 0 then patient has not any heart disease

```python
# splitting features and target
X = heart_disease_data.drop(columns = "target",axis = 1)
Y = heart_disease_data['target']
```

```python
# printing X and y values
print(X)
```

```python
# printing y value
print(Y)
```

Splitting data into training data and testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Logistric regression model training


```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

Model Evalution - Finding accuracy of training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
#printing accuracy score
print("Accurace of trained model = ",training_data_accuracy)
```

Finding accuracy of tested data


```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
# printing accuracy of tested data
print("Accurace of tested data = ",test_data_accuracy)
```

Predication from input data 

```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)

# change the input data to a numpy array
input_data_as_numpy_array= np.asarray(input_data)

# reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
